```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
import re
datata = pd.read_excel("./Python/datasets/printer_data_new.xlsx")
datata.head()
```

```python
datata = pd.DataFrame(datata)   # datata는 상품명, 가격 keyword 열이 사무기기 엑셀 파일 
                                # default값은 0 으로 두고, keyword에 해당 스펙이 존재할 경우 깂이 바뀜
datata['인쇄 해상도'] = int(0)  # 초기값 0 
cc = '인쇄 해상도: '
datata['인쇄 속도'] = int(0)    # 초기값 0
dd = '흑백 속도: '
ee = '컬러 속도: '

for i in range(0,5901):

    if( cc in datata.loc[i]['keyword']):            # keyword에 인쇄 해상도 스펙이 있을경우 temp에 값 저장
        start = datata.loc[i]['keyword'].find(cc)
        temp = datata.loc[i]['keyword'][start+8:start+13]
        temp = temp.split('dpi')[0]                 
        temp = temp.split(' x')[0]                  # 추출한 temp에서 단위제거 
        temp = temp.replace(',','')
        temp = re.findall('\d+',temp)
        temp[0] = int(temp[0])
        datata.at[i,'인쇄 해상도'] = temp[0]
    
    temp_arr=[]
    
    if( dd in datata.loc[i]['keyword']):            # keyword에 흑백 속도가 존재하면 temp배열에 값 저장
        start = datata.loc[i]['keyword'].find(dd)
        temp = datata.loc[i]['keyword'][start+4:start+10]
        temp = re.findall('\d+',temp)

        if(temp):
            temp[0] = int(temp[0])
            temp_arr.append(temp[0])
    if( ee in datata.loc[i]['keyword']):            # keyword에 컬러 속도가 존재하면 temp배열에 값 저장
        start = datata.loc[i]['keyword'].find(ee)
        temp = datata.loc[i]['keyword'][start+4:start+10]
        temp = re.findall('\d+',temp)

        if(temp):
            temp[0] = int(temp[0])
            temp_arr.append(temp[0])

        avg = sum(temp_arr,0)/len(temp_arr)         # temp배열의 평균값을 계산하여 인쇄 속도 열에 저장
        datata.at[i,'인쇄 속도'] = int(avg)

datata.head()
datata.to_excel('./printer_data_new2.xlsx', index = False)
```
