```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_excel('./datasets/복합기 크롤링 중복제거.xlsx')
```

```python
#user 평점 테이블 생성
item_user_table = pd.DataFrame(data=np.random.randint(1,6,size=(5901,15)), index=data['상품명'], columns=['대표님','팀장님','지은','시은','준영','도현','비둘기','세무사님','이룬아이티','아리오','후반기인턴1','후반기인턴2','후반기인턴3','정직원1'])
item_user_table
```

```python
#코사인 유사도로 유사도측정
printer_sim = cosine_similarity(item_user_table,item_user_table)
printer_sim_df = pd.DataFrame(data=printer_sim,index=item_user_table.index,columns=item_user_table.index)
printer_sim_df
```

```python
#사용자가 좋다고 평점매긴 것을 기준으로 상품추천.
def recommend_product(printer_sim_df, user_id, ori_ratings_df, num_recommendations=5):
    
    #item_user_table 평점 순서로 정렬
    sorted_ori_df = ori_ratings_df.sort_values(by=[user_id], ascending=False)
    product = sorted_ori_df.index[0]
    
    #가장높은 평점을 기준으로 10개 추천
    recommend = printer_sim_df[product].sort_values(ascending=False)[:10]
    
    return product, recommend
```

```python
product, recommend = recommend_product(printer_sim_df, '비둘기', item_user_table, 10)
```
