```python
ratings = pd.read_csv('./movie_data/ratings.csv')
movies = pd.read_csv('./movie_data/movies.csv')
```

```python
movies.rename(columns = {'id': 'movieId'}, inplace = True) #id를 movieId로 이름변경
```

```python
ratings_movies = pd.merge(ratings, movies, on = 'movieId') 
```

```python
ratings_movies.head(1)
```

```python
ratings_movies.shape
```

```python
data = ratings_movies.pivot_table('rating', index = 'userId', columns = 'title').fillna(0)
```

```python
data.head()
```

```python
data.shape
```

```python
data = data.transpose()
data.head(2)
```

```python
movie_sim = cosine_similarity(data, data)
print(movie_sim.shape)
```

```python
movie_sim_df = pd.DataFrame(data = movie_sim, index = data.index, columns = data.index)
```
